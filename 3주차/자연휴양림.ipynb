```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_excel('자연휴양림출력화일.xlsx')
df.head()
```

```python
map = folium.Map(
    location=[37.4965,126.9573],
    zoom_start=9,
    zoom_control=False,
    control_scale=True
)
map
```

```python
mc = MarkerCluster().add_to(map)
for n,row in df.iterrows():
    folium.Marker(location=[row['위도'],row['경도']],
                  popup = folium.Popup(row['휴양림명'],max_width=200)).add_to(mc)
map
```
